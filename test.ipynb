```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
@np.vectorize
def constant_function(x):
    return 1
```

```python
nx = 101
xl = np.linspace(start=-10, stop=-1, num=nx)
dx = 1e-2
xref = np.arange(start=-1 + dx, stop=1, step=dx)
xr = np.linspace(start=1, stop=10, num=nx)

x = np.concatenate([xl, xref, xr])
```

```python
plt.plot(x, np.abs(x), linewidth=0.5)
plt.plot(x, constant_function(x), linewidth=0.5)
```
