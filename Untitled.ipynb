```python
import numpy as np
import matplotlib.pyplot as plt

def casteljau_numpy(points, t):
    while len(points) > 1:
        points = (1 - t) * points[:-1] + t * points[1:]
    return points[0]

# Puntos de control como un arreglo NumPy de coordenadas (x, y)
control_points = np.array([
    [0, 0],
    [1, 2],
    [3, 3],
    [4, 1]
])

# Valores de parámetro t (entre 0 y 1)
t_values = np.linspace(0, 1, 100)  # Generamos 100 valores de t entre 0 y 1

# Calcular puntos en la curva Bézier para cada valor de t
curve_points = np.array([casteljau_numpy(control_points.copy(), t) for t in t_values])

# Extraer las coordenadas x e y de los puntos en la curva
x_curve = curve_points[:, 0]
y_curve = curve_points[:, 1]

# Graficar la curva Bézier
plt.figure(figsize=(8, 6))
plt.plot(x_curve, y_curve, label="Curva Bézier")
plt.scatter(control_points[:, 0], control_points[:, 1], c='red', label="Puntos de control")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curva Bézier')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def casteljau_nd(points, t):
    while len(points) > 1:
        points = (1 - t) * points[:-1] + t * points[1:]
    return points[0]

# Puntos de control como un arreglo NumPy de coordenadas n-dimensionales
control_points = np.array([
    [0, 0, 0],    # P0
    [1, 2, 1],    # P1
    [3, 3, 2],    # P2
    [4, 1, 3]     # P3
])

# Valores de parámetro t (entre 0 y 1)
t_values = np.linspace(0, 1, 100)

# Calcular puntos en la curva Bézier para cada valor de t
curve_points = np.array([casteljau_nd(control_points.copy(), t) for t in t_values])

# Extraer las coordenadas x, y, z de los puntos en la curva (solo para R3)
x_curve = curve_points[:, 0]
y_curve = curve_points[:, 1]
z_curve = curve_points[:, 2]

# Graficar la curva Bézier en R3
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_curve, y_curve, z_curve, label="Curva Bézier")
ax.scatter(control_points[:, 0], control_points[:, 1], control_points[:, 2], c='red', label="Puntos de control")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Curva Bézier en R3')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def casteljau_nd(points, t):
    while len(points) > 1:
        points = (1 - t) * points[:-1] + t * points[1:]
    return points[0]

# Puntos de control para formar una circunferencia
radius = 1
num_points = 4
theta = np.linspace(0, 2 * np.pi, num_points, endpoint=False)
control_points = np.column_stack((radius * np.cos(theta), radius * np.sin(theta)))

# Valores de parámetro t (entre 0 y 1)
t_values = np.linspace(0, 1, 100)

# Calcular puntos en la curva Bézier para cada valor de t
curve_points = np.array([casteljau_nd(control_points.copy(), t) for t in t_values])

# Coordenadas x, y de los puntos en la curva
x_curve = curve_points[:, 0]
y_curve = curve_points[:, 1]

# Graficar la curva Bézier como una circunferencia
plt.figure(figsize=(6, 6))
plt.plot(x_curve, y_curve, label="Curva Bézier (Circunferencia)")
plt.scatter(control_points[:, 0], control_points[:, 1], c='red', label="Puntos de control")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curva Bézier que forma una Circunferencia')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def casteljau_nd(points, t):
    while len(points) > 1:
        points = (1 - t) * points[:-1] + t * points[1:]
    return points[0]

# Puntos de control para formar una curva Bézier cúbica cerrada
control_points = np.array([
    [0, 0],
    [1, 2],
    [3, 3],
    [4, 1]
])

# Duplicar el primer punto para cerrar la curva
closed_control_points = np.vstack((control_points, control_points[0]))

# Valores de parámetro t (entre 0 y 1)
t_values = np.linspace(0, 1, 100)

# Calcular puntos en la curva Bézier cerrada para cada valor de t
curve_points = np.array([casteljau_nd(closed_control_points.copy(), t) for t in t_values])

# Coordenadas x, y de los puntos en la curva
x_curve = curve_points[:, 0]
y_curve = curve_points[:, 1]

# Graficar la curva Bézier cerrada
plt.figure(figsize=(8, 6))
plt.plot(x_curve, y_curve, label="Curva Bézier Cerrada")
plt.scatter(closed_control_points[:, 0], closed_control_points[:, 1], c='red', label="Puntos de control")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curva Bézier Cerrada')
plt.legend()
plt.grid(True)
plt.show()
```
